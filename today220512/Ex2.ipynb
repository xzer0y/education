```python
from sklearn.datasets import *
import pandas as pd
l_data = load_boston()
data = pd.DataFrame(l_data.data,columns=l_data.feature_names)
data['Y']=l_data.target
len(l_data.feature_names)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in range(3):
    k=list(l_data.feature_names[4*i:4+(4*i)])+['Y']
    sns.pairplot(vars=k,
             data=data)
    plt.show()
```
