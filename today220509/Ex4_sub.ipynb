```python
from tensorflow.keras.preprocessing.text import Tokenizer
from konlpy.corpus import kolaw
t_data=kolaw.open('constitution.txt')
```

```python
import kss
```

```python
from soynlp import DoublespaceLineCorpus
t_data=DoublespaceLineCorpus(t_data.name,iter_sent=True)
from soynlp.word import WordExtractor
for i in t_data:
    print(type(i))
    print(i)
```

```python
from konlpy.tag import Okt
tw=Okt()
d=[]
for i in t_data:
    s_data=tw.nouns(i)
    s_data=[i for i in s_data if len(i)>1]
    d.append(s_data)
d
```

```python
t_k=Tokenizer()
t_k.fit_on_texts(d)
t_k.index_word
```

```python
end_n_t=t_k.texts_to_sequences(d)
end_n_t
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
end_data = pad_sequences(end_n_t, padding='post', truncating='post')
end_data
```
