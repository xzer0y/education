머신 러닝 -> 모델 학습 -> 모델의 규칙은 상황에 따라 달라진다.<br>
머신 딥러닝 -> 머신러닝의 하위 -> 모델을 신경망을 이용하여 구축<br>
## data 가 가장 중요

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential # 모델 객체 생성(학습도구 만들기)
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

데이터 정리

```python
X=[1,2,3,4,5,6,7,8,9]
Y=[10,22,30,45,50,66,75,80,99]
```

모델

```python
m=Sequential()#모델생성
m.add(Dense(1,input_dim=1,activation='linear'))#모델 생성(모양 결정)
sgd = optimizers.SGD(lr=0.01)#오차수정규칙설정
m.compile(optimizer=sgd,loss='mse',metrics=['mse'])#모델의 학습규칙 설정
```

```python
m.fit(X,Y,epochs=2)#학습 방법=<지도학습>
```

```python
m.predict(X)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,m.predict(X),'b',X,Y,'k.')
```

학습이란 가중치의 갱신
- loss => 작을수록 좋다
- 정확도-> 클수록 좋다
- y=wx+b

데이터 정리

```python
X=[-40,-30,-20,-10,0,10,20,30,40]
Y=[0,0,0,0,1,1,1,1,1]
```

모델 생성

```python
m=Sequential()
m.add(Dense(1, input_dim=1,activation='sigmoid'))
m.compile(optimizer=optimizers.SGD(lr=0.01),loss='binary_crossentropy'
          ,metrics=['binary_accuracy'])
```

```python
m.fit(X,Y,epochs=200)
```

```python
m.predict(X)
```

```python
plt.plot(X,m.predict(X),'b',X,Y,'k.')
```
